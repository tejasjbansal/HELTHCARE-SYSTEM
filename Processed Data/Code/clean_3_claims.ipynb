```python
# Importing the processing Libray..
import numpy as np
import pandas as pd
from datetime import datetime
```

```python
# Count the total Null values
df.isnull().sum()
```

```python
# Replace the null values
df['Claim_Or_Rejected'].mask(df['Claim_Or_Rejected'] == 'NaN', 'N', inplace=True)
```

```python
df.head(60)
```

```python
# Check the duplicates
df.duplicated()
```

```python
# Drop duplicates
df.drop_duplicates(subset=None, keep='first', inplace=False, ignore_index=False)
```

```python
# Check Subscriber Id not less than 9 digit
count = 0
for i in df['SUB_ID'].values:
    
    if len(i)<9:
        df.drop([count], axis=0, inplace=True)
    elif len(str(i))>10:
        df.drop([count], axis=0, inplace=True)
    count = count+1
```

```python
# Check Patient Id not less than 6 digit
count = 0
for i in df['patient_id'].values:
    
    if len(str(i))<6:
        df.drop([count], axis=0, inplace=True)
    elif len(str(i))>6:
        df.drop([count], axis=0, inplace=True)
    count = count+1
```
